## convert the datatype of rate column

```python
def handlerate(value):
    value = value.split('/')
    value = value[0]
    return float(value)

df['rate'] = df['rate'].apply(handlerate)
print(df.head())
```

## conclusion - majority of rsturants fall in dining category

```python
import matplotlib.pyplot as plt
import pandas as pd

# Group data
grouped_data = df.groupby('listed_in(type)')['votes'].sum()

# Convert to DataFrame
result = grouped_data.reset_index()

# Plot
plt.figure(figsize=(8, 5))
plt.plot(result['listed_in(type)'], result['votes'], marker='o')

plt.xlabel("Type of Restaurant")
plt.ylabel("Votes")
plt.title("Votes by Restaurant Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### conclusion- dining resturant has received maximum votes 

```python
plt.hist(df['rate'],bins=50)
plt.title("Rating distibution")
plt.show()
```

### conclusion- majority of resturant receive rating 3.5 to 4

### Which mode Receives highest rating?

```python
sns.boxplot(x= 'online_order', y= 'rate',data=df)
```

### conclusion- offline order receive lower rating in comparison to online rating

```python

pivot_table = df.pivot_table(
    index='listed_in(type)',
    columns='online_order',
    aggfunc='size',
    fill_value=0
)

sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')

plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

### Conclusion- Dining resturants accept offline orders ,whereas cafes primarily receive online orders. This suggests that clients prefer orders in person at resturants, but prefer online at cafes
